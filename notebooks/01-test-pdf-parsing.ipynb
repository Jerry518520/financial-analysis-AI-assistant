```python
# 导入 PyMuPDF 库，它的名字叫 fitz
import fitz
import os
# 我们使用 os.path.join 来构建一个更可靠的跨平台路径
# ".." 代表返回上一级目录
pdf_path = os.path.join("..", "example_report.pdf") 

# 打开 PDF 文件
try:
    doc = fitz.open(pdf_path)
    print(f"文件名: {pdf_path}")
    print(f"总页数: {doc.page_count}")
    print("-" * 20)
    first_page = doc.load_page(0) 
    text = first_page.get_text()
    
    print("第一页的文本内容 (前300个字符):\n")
    print(text[:300]) 
    doc.close()

except Exception as e:
    print(f"处理 PDF 时出错: {e}")
    print("\n请确保你已经在项目的根目录下放置了名为 'example_report.pdf' 的文件。")
```

```python
# ---------------------------------------------------------
# MVP 验证步骤 2: 尝试提取表格
# ---------------------------------------------------------
import fitz # PyMuPDF
import pandas as pd # 用于展示表格

# 1. 重新加载文档（为了确保环境干净）
doc = fitz.open("../example_report.pdf")

# 2. 我们假设第 5 页通常有财务数据（你可以根据实际PDF改成其他页码，比如 0, 1, 2...）
# 提示：找一页肉眼看过去有“表格线”或者“整齐数据”的页面
target_page_index = 188  # 注意：计算机从0开始计数，4代表第5页
page = doc.load_page(target_page_index)

# 3. 呼叫 PyMuPDF 的表格查找功能
print(f"正在分析第 {target_page_index + 1} 页的表格...")
tables = page.find_tables()

# 4. 看看找到了没
if tables.tables:
    print(f"🎉 成功！在这一页发现了 {len(tables.tables)} 个表格。")
    
    # 把第一个表格转换成 Pandas DataFrame 显示出来
    df = tables[0].to_pandas()
    print("第一个表格的内容预览：")
    display(df) # 在 Jupyter 里直接漂亮地显示表格
else:
    print("⚠️ 这一页没发现表格。请尝试修改 target_page_index 换一页试试。")

doc.close()
```
